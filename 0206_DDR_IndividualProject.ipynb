```python
from bs4 import BeautifulSoup
import requests
import time
import os
```

```python
url = 'https://sfbay.craigslist.org/search/zip?sort=date#search=1~gallery~0~0'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36 Edg/120.0.0.0'}
page = requests.get(url,headers)
soup = BeautifulSoup(page.content,'html.parser')
```

```python
# Initialize an empty set to store unique hrefs
unique_hrefs = set()

# find the link within the <ol>
ol_tags = soup.find_all('ol')
for ol in ol_tags:
    links = ol.find_all('a')
    for link in links:
        href = link.get('href')
        if href and href not in unique_hrefs: # check if it is unique
          unique_hrefs.add(href)
          if len(unique_hrefs) == 250: # check to meet the limit
            break

# Convert the set to a list and print
hrefs_list = list(unique_hrefs)
print(hrefs_list)
```

```python
# Directory to save the HTML files
save_directory = 'listings_html'
os.makedirs(save_directory, exist_ok=True)

# Initialize the counter for successfully saved listings
success_count = 0

# Iterate over the URLs
for href in hrefs_list:
    try:
        response = requests.get(href)
        # take a pause
        time.sleep(7)

        # Check if the request was successful
        if response.status_code == 200:
            # Get the content of the response
            content = response.text

            # Extract the listing ID from the URL
            listing_id = href.split('/')[-1].replace('.html', '')

            # Define the file name using the listing ID
            file_name = f"{listing_id}.html"

            # Define the file path
            file_path = os.path.join(save_directory, file_name)

            # Write the content to the file
            with open(file_path, 'w', encoding='utf-8') as file:
                file.write(content)
                success_count += 1  # Increment the success counter
        else:
            print(f"Failed to fetch {href}: Status code {response.status_code}")
    except requests.RequestException as e:
        print(f"Request failed for {href}: {str(e)}")

# Print the total number of successfully saved listings
print(f"All listings processed. Successfully saved {success_count} listings.")
```
